# Multimodal Mental Health Scoring — Data Exploration

This notebook helps you explore your datasets before training.

```python
import sys
sys.path.insert(0, '..')
import config
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

## 1. Text Dataset

```python
text_df = pd.read_csv(config.TEXT_CONFIG['raw_csv'])
print(f'Shape: {text_df.shape}')
print(f'\nLabel distribution:\n{text_df["label"].value_counts()}')
text_df.head()
```

```python
text_df['text_length'] = text_df['text'].str.len()
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
text_df['label'].value_counts().plot.bar(ax=axes[0], color=['#10b981', '#ef4444'])
axes[0].set_title('Label Distribution')
axes[0].set_xticklabels(['Normal', 'Distress'], rotation=0)
text_df.boxplot(column='text_length', by='label', ax=axes[1])
axes[1].set_title('Text Length by Label')
plt.tight_layout()
plt.show()
```

## 2. Questionnaire Dataset

```python
quest_df = pd.read_csv(config.QUESTIONNAIRE_CONFIG['raw_csv'])
print(f'Shape: {quest_df.shape}')
print(f'\nLabel distribution:\n{quest_df["risk_level"].value_counts()}')
quest_df.describe()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
feature_cols = [c for c in quest_df.columns if c != 'risk_level']
corr = quest_df[feature_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
ax.set_title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
